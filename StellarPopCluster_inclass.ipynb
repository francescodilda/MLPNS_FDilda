```python
import pandas as pd
import pylab as pl
import sklearn as skl
from sklearn import preprocessing
import scipy as sp
from scipy import spatial
import numpy as np
from sklearn import cluster
```

```python
url = "https://raw.githubusercontent.com/fedhere/MLPNS_FBianco/main/clustering/colormag_NGC2808.csv"
cmag = pd.read_csv(url, header = None)
```

```python
cmag.describe()
```

```python
pl.figure(figsize = (5, 5));
ax = cmag.plot(x = 0, y = 1, kind='scatter');
ax.set_xlabel("B-V (color)");
ax.set_ylabel("V mag)");
ax.axis("equal")
ax.invert_yaxis();
```

```python
cmag_scaled = preprocessing.scale(cmag, axis=0)
```

```python
ax = pl.figure(figsize = (5, 5)).add_subplot(111);
ax.plot(cmag_scaled[:, 0],cmag_scaled[:, 1], '.');
ax.set_xlabel("B-V (color)");
ax.set_ylabel("V mag)");
ax.axis("equal")
ax.invert_yaxis();
```

```python
#adesso possiamo applicare il metodo k-means
cluster.KMeans?
```

```python
model = cluster.KMeans(n_clusters = 3, random_state = 302)
```

```python
model.fit(cmag_scaled)
```

```python
model.labels_
```

```python
model.cluster_centers_
```

```python
model.inertia_
```

```python
ax = pl.figure(figsize = (5, 5)).add_subplot(111);
for i in range(3):
  ax.plot(cmag_scaled[model.labels_ == i, 0],cmag_scaled[model.labels_ == i, 1], '.');
  ax.plot(model.cluster_centers_[i, 0], model.cluster_centers_[i, 1], 'kx');
  ax.set_xlabel("B-V (color)");
  ax.set_ylabel("V mag)");
  ax.axis("equal")
  ax.invert_yaxis();

  #Riconosco il problema  che la forma dei vari cluster è sempre tendenzialmente sferica per com'è fatto l'algoritmo
```

```python
inertias = []
for nc in range(2, 10):
  model = cluster.KMeans(n_clusters = nc, random_state = 302).fit(cmag_scaled);
  inertias.append(model.inertia_);

pl.plot(inertias)
```

```python
modeldbs = cluster.DBSCAN(eps = 0.001).fit(cmag_scaled)         #non usare senza una domain knowledge che ti permette di definire una scale di distanze sensate
```

```python
ax = pl.figure(figsize = (5, 5)).add_subplot(111);
for i in np.unique(modeldbs.labels_):
  ax.plot(cmag_scaled[modeldbs.labels_ == i, 0],cmag_scaled[modeldbs.labels_ == i, 1], '.');

ax.set_xlabel("B-V (color)");
ax.set_ylabel("V mag)");
ax.axis("equal")
ax.invert_yaxis();
```

```python
for e in np.arange(0.001, 1, 0.001):
  modeldbs = cluster.DBSCAN(eps = e).fit(cmag_scaled)
  if modeldbs.labels_.max() > 0:
    print(e, modeldbs.labels_.max())
```

```python
modeldbs = cluster.DBSCAN(eps = 0.2).fit(cmag_scaled)

ax = pl.figure(figsize = (5, 5)).add_subplot(111);
for i in np.unique(modeldbs.labels_):
  ax.plot(cmag_scaled[modeldbs.labels_ == i, 0],cmag_scaled[modeldbs.labels_ == i, 1], '.');
  print(i)
ax.set_xlabel("B-V (color)");
ax.set_ylabel("V mag)");
ax.axis("equal")
ax.invert_yaxis();
```
